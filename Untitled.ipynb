def export_table_to_csv(connection, table_name, csv_file_name, batch_size=1000):
    cursor = connection.cursor()
    query = f"SELECT * FROM {table_name}"
    cursor.execute(query)
    with open(csv_file_name, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        # Write headers
        csv_writer.writerow([i[0] for i in cursor.description])
        # Fetch and write data in batches
        while True:
            records = cursor.fetchmany(batch_size)
            if not records:
                break
            csv_writer.writerows(records)
    cursor.close()
    print(f"Data exported to {csv_file_name} successfully")

```python
import pandas as pd
import numpy as np

# Example dataset
df_posts = pd.DataFrame({
    'influencer_id': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'post_id': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'likes': [100, 1000000, 150, 200, 250, 1000000, 300, 100, 100],
    'created_time': pd.to_datetime(['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-01', '2022-01-02', '2022-01-03', '2022-01-01', '2022-01-02', '2022-01-03'])
})

# Define outliers based on likes
threshold_factor = 10  # Example factor to identify outliers
df_posts['avg_likes'] = df_posts.groupby('influencer_id')['likes'].transform('mean')
df_posts['outlier'] = np.where(df_posts['likes'] > threshold_factor * df_posts['avg_likes'], 1, 0)

# Define pre and post periods based on outlier post dates
outlier_dates = df_posts[df_posts['outlier'] == 1].groupby('influencer_id')['created_time'].min().reset_index()
outlier_dates.columns = ['influencer_id', 'outlier_date']

df_posts = df_posts.merge(outlier_dates, on='influencer_id', how='left')
df_posts['period'] = np.where(df_posts['created_time'] < df_posts['outlier_date'], 'pre', 'post')
```

```python
import statsmodels.api as sm

# Create dummy variables for treatment and post periods
df_posts['Treatment'] = df_posts['outlier']
df_posts['Post'] = df_posts['period'].apply(lambda x: 1 if x == 'post' else 0)
df_posts['Treatment_Post'] = df_posts['Treatment'] * df_posts['Post']

# Define the dependent variable (e.g., likes)
y = df_posts['likes']

# Define the independent variables
X = df_posts[['Treatment', 'Post', 'Treatment_Post']]
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# View the results
print(model.summary())
```

## Counting pre and post article before viral article

```python
import numpy as np
import pandas as pd

# Function to calculate standardized mean differences
def calculate_smd(df, treatment_col, covariate_cols):
    smd = {}
    treatment_group = df[df[treatment_col] == 1]
    control_group = df[df[treatment_col] == 0]
    for col in covariate_cols:
        mean_treatment = treatment_group[col].mean()
        mean_control = control_group[col].mean()
        pooled_std = np.sqrt((treatment_group[col].std() ** 2 + control_group[col].std() ** 2) / 2)
        smd[col] = (mean_treatment - mean_control) / pooled_std
    return pd.Series(smd)

# Specify the columns
treatment_col = 'treatment'
covariate_cols = ['followerCount', 'followingCount', 'gender_0', 'gender_1']

# Calculate SMD for matched data
smd_matched = calculate_smd(df_matched, treatment_col, covariate_cols)
print(smd_matched)
```

```python
# Exoginity Check
residuals=model.resid
df_matched['residuals']=residuals
correlation_matrix_with_residuals = df_matched[['residuals', 'voteupCount', 'followerCount', 'followingCount', 'gender_0', 'gender_1', 'treatment', 'post_viral']].corr()

# Display the correlation matrix
print(correlation_matrix_with_residuals)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap of the correlation matrix including residuals
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_with_residuals, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Including Residuals')
plt.show()
```
